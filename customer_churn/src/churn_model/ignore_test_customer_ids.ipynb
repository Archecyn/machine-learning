```python
import logging
from datetime import datetime
from pyspark.sql import DataFrame as SparkDataFrame
from pyspark.sql.functions import col

logger = logging.getLogger("churn_model")


def ignore_test_customer_ids(
        df: SparkDataFrame, 
        *,
        test_customer_ids: list = None
    ) -> SparkDataFrame:
    """Flag test customer IDs with a boolean column ``test_customer_ids``."""

    logger.info(f"ignore_test_customer_ids started at {datetime.now()}")
    result = df.withColumn("test_customer_ids", col("customer_ids").isin(test_customer_ids))
    logger.info(f"ignore_test_customer_ids finished at {datetime.now()}")

    return result
```
