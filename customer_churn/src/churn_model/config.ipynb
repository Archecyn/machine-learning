```python
"""
Configuration module for the customer churn prediction project.
This is to access all values via dot notation, e.g. ``config.default.schema``.
"""

class _Section:
    """Simple namespace that supports dot-access on its attributes."""
    def __init__(self, **kwargs):
        for k, v in kwargs.items():
            setattr(self, k, v)

    def __repr__(self):
        attrs = ", ".join(f"{k}={v!r}" for k, v in self.__dict__.items())
        return f"_Section({attrs})"


class Config:
    default = _Section(
        schema="customer_churn",
        logging_volume="logging",
        base_table="customer.customers",
        target_table="customer.churn_predictions",
        model_name="customer_churn_model",
        artifact_path="customer_churn_artifacts",
    )

    data_columns = _Section(
        customer_id="customer_id",
        label="churn_flag",
        event_date="event_date",
        churn_window_days=30,
        prediction_threshold=0.5,
        base_hazard_rate=0.02,
    )

    feature_groups = _Section(
        demographic=["age", "gender", "location"],
        behaviour=["logins", "session_length", "purchase_count"],
        financial=["monthly_spend", "credit_score", "payment_history"],
    )

    sampling = _Section(
        class_weight="balanced",
        sampling_strategy="undersample",
        oversample_ratio=1.0,
    )

    validation = _Section(
        train_ratio=0.8,
        cv_folds=5,
        stratify=True,
    )

    monitoring = _Section(
        drift_threshold=0.1,
        alert_email="archecyn@users.noreply.github.com",
    )

    model_parameters = _Section(
        learning_rate=0.01,
        regularisation=0.1,
        regularisation_type="l2",
        tree_depth=6,
        n_estimators=100,
        min_samples_leaf=10,
        max_features="sqrt",
        early_stopping_rounds=10,
        objective="binary:logistic",
    )

    test_customer_ids = _Section(
        customer_ids=[
            'CID100960237',
            'CID108250834',
            'CID109059419',
            'CID169038206',
        ],
    )

config = Config()
```
